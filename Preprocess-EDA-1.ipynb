4. Điểm tổng kết should be of type float64, not string. The reason that it is currenly string is because some special grades got mixed up as well ('CT', 'VT', 'RT', 'IH','HT', 'KD', 'DT' and numeric >10 scores)

```python
print(len(df["F_TENMHVN"].unique()))
print(len(df["F_MAMH"].unique()))
print("Giải tích 1 has 3 subject codes: "+ str(df[df['F_TENMHVN'] == "Giải tích 1"]["F_MAMH"].unique()))
print("Những ngLý cơbản cn M-Lê has the subject code: "+ str(df[df['F_TENMHVN'] == "Những ngLý cơbản cn M-Lê"]["F_MAMH"].unique()))
print("Những NgLý cơbản CN M-Lê has the subject code: "+ str(df[df['F_TENMHVN'] == "Những NgLý cơbản CN M-Lê"]["F_MAMH"].unique()))
```

```python
def draw_his_1_subject(subject_name: str, df):
  fig = px.histogram(df[df.F_TENMHVN == subject_name], x= "TKET")
  fig.show()
```

```python
draw_his_1_subject("Mạch điện tử", df_e) # See the score distribution of one particular subject
```

```python
def draw_his_1_student(student_ID: int, df):
  fig = px.bar(df[df.MASV == student_ID], y='TKET', x='F_TENMHVN', text = "NHHK", color = "NHHK")
  fig.update_layout(barmode = 'group')
  fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
  #fig.add_hrect(y0=0, y1=4.99999, line_width=0, fillcolor="red", opacity=0.3)
  fig.show()
```

```python
draw_his_1_student(7287845, df_e)
```

This led to his decision to take both these two subjects again in the summer semester 153.

Understandably, semester 161, with 12 different classes, is too much for him to retake anymore subjects.

And for semester 162, for subject **Vi xử lý** and **Bảo vệ relay trong htd** which he failed during semester 161, he immediately retake them again in 162, as well as two subjects Kỹ thuật số and Giải tích mạch, which he failed in 151.

In short, recommending subjects that students have yet to pass is a valid suggestion, due to their desire to finish what they have started while their knowledge on the subject is still there.
