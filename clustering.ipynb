```python
import pandas as pd
import numpy as np
df = pd.read_csv('diem_clean_with_0.csv', dtype={'F_MAMH': str, "NHHK": str})
```

```python
# TKET >= 9.0: Exellent
# 8.0 <= TKET < 9.0: Good
# 7.0 <= TKET < 8.0: Average
# 5.0 <= TKET < 7.0: Meh
# TKET < 5.0: Failed

conditions = [
    (df['TKET'] >= 9.0),
    (df['TKET'] >= 8.0) & (df['TKET'] < 9.0),
    (df['TKET'] >= 7.0) & (df['TKET'] < 8.0),
    (df['TKET'] >= 5.0) & (df['TKET'] < 7.0),
    (df['TKET'] < 5.0)
]

values = ['Exellent', 'Good', 'Average', 'Meh', 'Failed']

df["TKET_RANK"] = np.select(conditions, values)

```

```python
import pandas as pd
from sklearn.decomposition import PCA

df = pd.read_csv('diem_clean_with_0_with_rank.csv', dtype={'F_MAMH': str, "NHHK": str})
```

```python
#df.F_TENMHVN.value_counts().to_csv('subject_count.csv')
```

```python
subjects = "Đại số tuyến tính"
NHHK = "162"
```

```python
features = ["MASV1","BT", "BTLDA", "TN", "KT", "THI", "TKET", "TKET_RANK"]
filtered_df = df[(df.F_TENMHVN == subjects) & (df.NHHK == NHHK)][features]
filtered_df = filtered_df[filtered_df != 13].dropna()
```

```python
filtered_df
```

```python
df_masv = filtered_df["MASV1"]
df_masv.shape
```

```python
filtered_df = filtered_df.drop(["MASV1"], axis=1)
```

```python
filtered_df.shape
```

```python
df_labels = filtered_df["TKET_RANK"]
df_labels.shape
```

```python
filtered_df = filtered_df.drop(["TKET_RANK"], axis=1)
```

```python
filtered_df.BTLDA.value_counts()
```

# K-Means Clustering

```python
# apply kmeans on filtered_df
from sklearn.cluster import KMeans
kmeans_data = KMeans(n_clusters=4, random_state=0).fit_transform(filtered_df)
kmeans_data
```

# PCA

```python
# reset index all df
df_masv = df_masv.reset_index(drop=True)
filtered_df = filtered_df.reset_index(drop=True)
df_labels = df_labels.reset_index(drop=True)
```

```python
import numpy as np
```

```python
# Train PCA
pca = PCA(n_components=2)
transform_array = pca.fit_transform(filtered_df)
transform_df = pd.DataFrame(data=transform_array, columns=["PC1", "PC2"])
print(pca.explained_variance_ratio_)
np.sum(pca.explained_variance_ratio_)
```

```python
type(transform_df)
```

```python
transform_df.info()
```

```python
# concat transform_df with df_labels and filtered_df and df_masv
final_df = pd.concat([transform_df, df_labels, filtered_df, df_masv], axis=1)

```

```python
final_df.info()
```

```python
import plotly.express as px
px.scatter(final_df, x="PC1", y="PC2", color=df_labels, hover_data={"PC1": False, "PC2": False, "TKET_RANK": False, "BT": True,
"BTLDA": True, "TN": True, "KT": True, "THI": True, "MASV1": True} )
```

# TSNE

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
transform_df_tsne = tsne.fit_transform(filtered_df)
```

```python
# concat transform_df_tsme with df_labels and filtered_df and df_masv
final_df_tsne = pd.concat([pd.DataFrame(data=transform_df_tsne, columns=["PC1", "PC2"]), df_labels, filtered_df, df_masv], axis=1)
```

```python
import plotly.express as px
px.scatter(final_df_tsne, x="PC1", y="PC2", color=df_labels, hover_data={"PC1": False, "PC2": False, "TKET_RANK": False, "BT": True,
"BTLDA": True, "TN": True, "KT": True, "THI": True, "MASV1": True} )
```
